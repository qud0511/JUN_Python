```python
# MySQL과 Python 연동하기
conn = pymysql.connect(host='localhost', user='root', password='0000', 
db = 'shoppingmall', charset='utf8')

cur = conn.cursor() # 2) 커서 객체 생성
cur.execute('select	* from user_table') # 3) 쿼리 실행
user_table_rows = cur.fetchall() # 4) fetchall() : 모든 검색 데이터 가져오기
print(user_table_rows) # for row in rows: print(row)와 동일

user_table_df = pd.DataFrame(user_table_rows) # DataFrame 형태로 변환
print(user_table_df)

# conn.commit()
# conn.close()
```

```python
cur.execute('''select u.userName, b.prodName, u.addr, concat(u.mobile1, u.mobile2)
from user_table as u inner join buy_table as b
on u.userID = b.userID;''')

num2_1 = cur.fetchall()
print(num2_1)

pd.DataFrame(num2_1)
```

```python
cur.execute('''select u.userName, b.prodName, u.addr, concat(u.mobile1, u.mobile2)
from user_table as u inner join buy_table as b
on u.userID = b.userID
where b.userID = 'KYM';''')

num2_2 = cur.fetchall()
print(num2_2)

pd.DataFrame(num2_2)
```

```python
cur.execute('''select u.userName, b.prodName, u.addr, concat(u.mobile1, u.mobile2)
from user_table as u inner join buy_table as b
on u.userID = b.userID
where b.userID = 'KYM';''')

num2_2 = cur.fetchall()
print(num2_2)

pd.DataFrame(num2_2)
```

```python
cur.execute('''select u.userID, u.userName, b.prodName, u.addr, concat(u.mobile1, u.mobile2) as '연락'
from user_table as u inner join buy_table as b
on u.userID = b.userID
order by b.userID;''')

num3 = cur.fetchall()
print(num3)

pd.DataFrame(num3)
```

```python
cur.execute('''select distinct u.userID, u.userName, u.addr
from user_table as u inner join buy_table as b
on u.userID = b.userID
order by b.userID;''')

num4 = cur.fetchall()
print(num4)

pd.DataFrame(num4)
```

```python
cur.execute('''select distinct userID, userName, addr, concat(mobile1, mobile2) as '연락'
from user_table
where addr in ('경북', '경남');''')

num5 = cur.fetchall()
print(num5)

pd.DataFrame(num5)
```
