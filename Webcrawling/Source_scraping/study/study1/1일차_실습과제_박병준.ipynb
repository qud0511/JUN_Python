### for문

```python
# tombstone-container, short-desc, temp 모두 찾기
container_list = soup.find_all('div', {'class' : 'tombstone-container'})
container_list
# soup.find_all('p', {'class' : 'period-name'}).text => find-all에서 text만 가져올 수 없음.
```

```python
container_list[0]
```

```python
for i in range(len(container_list)): # container_list의 [0] ~ [8] => range(9)와 같음.
    print('[Period]', container_list[i].find('p', {'class' : 'period-name'}).text)
    print('[Short desc]', container_list[i].find('p', {'class' : 'short-desc'}).text)
    print('[Temperature]', container_list[i].find('p', {'class' : 'temp'}).text)
    print('[Image desc]', container_list[i].find('img')['title'])
    print()
```

## def scraping_use_find(html)함수 만들기

```python
def scraping_use_find(soup):
    tcl = soup.find_all('div', {'class' : 'tombstone-container'}) # tombstone_container_list
    
    print('[Find 함수이용]')
    print('-'*80)
    for i in tcl: # i는 tcl[0] ~ 끝까지
        print('[Period]', i.find('p', {'class' : 'period-name'}))
        print('[Short desc]', i.find('p', {'class' : 'short-desc'}))
        print('[Temperature]', i.find('p', {'class' : 'temp'}))
        print('[Image desc]', i.find('img')['title'])
        print('-'*80)
```

```python
scraping_use_find(soup)
```
