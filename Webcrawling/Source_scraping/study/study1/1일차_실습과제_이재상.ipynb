```python
# 모듈 로딩
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# 사용자 정의 함수
# 스크레이핑 함수 (find())
def scraping_use_find(html):
    h = urlopen(html)
    bs = BeautifulSoup(h.read(), 'html.parser')
    a = bs.select('div.tombstone-container')
    for link in a:
        a1 = link.find('p', {'class' : 'period-name'}).text
        a2 = link.find('p', {'class' : 'short-desc'}).text
        a3 = link.find('p', {'class' : 'temp'}).text
        a4 = link.find('img')
        a4 = a4['title']
        print('[Period] : {0}\n[Short desc] : {1}\n[Temperature] : {2}\n[Image desc] : {3}\n'.format(a1, a2, a3, a4))

# 스크레이핑 함수 (select_one())
def scraping_use_select(html):
    h = urlopen(html)
    bs = BeautifulSoup(h.read(), 'html.parser')
    a = bs.select('div.tombstone-container')
    for link in a:
        a1 = link.select_one('p.period-name').text
        a2 = link.select_one('p.short-desc').text
        a3 = link.select_one('p.temp').text
        a4 = link.select_one('p>img')
        a4 = a4['title']
        print('[Period] : {0}\n[Short desc] : {1}\n[Temperature] : {2}\n[Image desc] : {3}\n'.format(a1, a2, a3, a4))

```

```python
scraping_use_find('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YtkPH3ZByUk')
scraping_use_select('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YtkPH3ZByUk')
```
