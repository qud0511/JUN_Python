```python
def scraping_use_find(html):
    from urllib.request import urlopen
    from bs4 import BeautifulSoup
    
    html = urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytjs5-xByYs')
    bs = BeautifulSoup(html.read(), 'html.parser')
    bs_1=bs.find('div', {'id':'seven-day-forecast'})
    
    bs_period=bs_1.find_all('p', {'class':'period-name'})
    bs_sd=bs_1.find_all('p', {'class':'short-desc'})
    bs_temp=bs_1.find_all('p', {'class':'temp'})
    bs_img=bs_1.find_all('img')
    
    print('National Weather Service Scraping')
    print('---------------------------------')
    print('[find 함수 이용]')
    print('------------------------------------------------------------------')

    for i in range(len(bs_period)):
        print('[Period]: ', bs_period[i].text)
        print('[Short desc]: ', bs_sd[i].text)
        print('[Temperature]: ', bs_temp[i].text)
        print('[Image desc]: ', bs_img[i]['title'])
        print('------------------------------------------------------------------')
```

```python
def scraping_use_select(html):
    from urllib.request import urlopen
    from bs4 import BeautifulSoup

    html = urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytjs5-xByYs')
    bs = BeautifulSoup(html.read(), 'html.parser')
    
    bs_s_1=bs.select_one('div#seven-day-forecast')
    
    bs_s_1_period=bs_s_1.select('p.period-name')
    bs_s_1_sd=bs_s_1.select('p.short-desc')
    bs_s_1_temp=bs_s_1.select('p.temp')
    bs_s_1_img=bs_s_1.select('img')

    print('National Weather Service Scraping')
    print('---------------------------------')
    print('[select 함수 이용]')
    print('------------------------------------------------------------------')

    for i in range(len(bs_s_1.select('p.period-name'))):
        print('[Period]: ', bs_s_1_period[i].text)
        print('[Short desc]: ', bs_s_1_sd[i].text)
        print('[Temperature]: ', bs_s_1_temp[i].text)
        print('[Image desc]: ', bs_s_1_img[i]['title'])
        print('------------------------------------------------------------------')
```

```python
scraping_use_find('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytjs5-xByYs')
```

```python
scraping_use_select('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytjs5-xByYs')
```
