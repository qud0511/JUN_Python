# 데이터 크롤링과 정제

## 과제 1번: Weather Service

- 일주일간의 일기예보 내용을 화면에 출력
- 총 8일간의 일기예보 내용을 스크레이핑

- 출력 항목 (4개 항목)  
        – period-name
        – short-desc
        – temp
         – <img title='...'>

### def scraping_use_find(html)
– find(), find_all() 함수를 이용하여 스크레이핑

```python
from urllib.request import urlopen
from  bs4 import BeautifulSoup

def scraping_use_find(html):
    data=BeautifulSoup(html.read(), 'html.parser')
    print('National Weather Service Scraping')
    print('----------------------------------\n[find 함수 이용]\n--------------------------------------------------------------------------------')
    divs=data.find_all('div',{'class':'tombstone-container'})
    for div in divs:
        print(f"[Period]: {div.find('p',{'class':'period-name'}).text}")
        print(f"[Short desc]: {div.find('p',{'class':'short-desc'}).text}")
        print(f"[Temperature]: {div.find('p',{'class':'temp'}).text}")
        img_tag=div.find('img')
        print(f"[Image desc]: {img_tag.attrs['title']}")
        print('--------------------------------------------------------------------------------')
    

html=urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YtjwPnZBxPY')    

scraping_use_find(html)
```

### def scraping_use_select(html)
– select(), select_one() 함수를 이용하여 스크레이핑

```python
from urllib.request import urlopen
from  bs4 import BeautifulSoup

def scraping_use_select(html):
    data=BeautifulSoup(html.read(), 'html.parser')
    print('National Weather Service Scraping')
    print('----------------------------------\n[select 함수 이용]\n--------------------------------------------------------------------------------')
    datas=data.select_one('div#seven-day-forecast')
    a=[]  # period-name 담을 리스트
    b=[]  # short-desc 담을 리스트  
    c=[]  # temp 담을 리스트
    e=[]  # img_title 담을 리스트
    for i in datas.select('p.period-name'):
        a.append(i.text)
    for i in datas.select('p.short-desc'):
        b.append(i.text)
    for i in datas.select('p.temp'):
        c.append(i.text)
    d=datas.select('img')
    for i in d:
        e.append(i['title'])
    for j in range(len(a)):
        print(f"[Period]: {a[j]}")
        print(f"[Short desc]: {b[j]}")
        print(f"[Temperature]: {c[j]}")
        print(f"[Image desc]: {e[j]}")
        print('--------------------------------------------------------------------------------')
    

html=urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YtjwPnZBxPY')    

scraping_use_select(html)
```
