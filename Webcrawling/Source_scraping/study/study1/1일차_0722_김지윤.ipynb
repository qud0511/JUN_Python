```python
from urllib.request import urlopen
from bs4 import BeautifulSoup


html=urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YtpVkHZByUk')
# bs=BeautifulSoup(html.read(), 'html.parser')
soup=BeautifulSoup(html.read(), 'html.parser')
```

```python
# find 사용
a=soup.find_all('div', {'class':'tombstone-container'})
print('[find 함수 이용]')
print('-'*70)
for i in a:
    P=i.find_all('p', {'class':'period-name'})
    S=i.find_all('p', {'class':'short-desc'})
    T=i.find_all('p', {'class':'temp'})
    I=i.find_all('img')
    for i in P:
        print(f'[period]: {i.text}')
    for i in S:
        print(f'[short-desc]: {i.text}')
    for i in T:
        print(f'[temp]: {i.text}')
    for i in I:
        print('[Image desc]: '+ i['title'])
    print('-'*70)
    
    
 
```

```python
print('[select 함수 이용]')
print('-'*70)
a=soup.select('div.tombstone-container')
for i in a:
    P=i.select('p.period-name')
    S=i.select('p.short-desc')
    T=i.select('p.temp')
    I=i.select('img')
    for i in P:
        print(f'[period]: {i.text}')
    for i in S:
        print(f'[short-desc]: {i.text}')
    for i in T:
        print(f'[temp]: {i.text}')
    for i in I:
        print(f'[Image desc]: '+ i['title'])
    print('-'*70)
```
