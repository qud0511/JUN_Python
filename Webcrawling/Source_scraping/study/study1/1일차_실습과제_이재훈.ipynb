과제 내용

§ 일주일간의 일기예보 내용을 화면에 출력

• 총 8일간의 일기예보 내용을 스크레이핑

• 출력 항목 (4개 항목)
    – period-name

    – short-desc

    – temp
    
    – <img title=“...”>	

§ 2개의 함수를 작성

• 출력 내용은 동일하지만, 아래 2개의 함수를 각각 구현하시오.

• def	scraping_use_find(html)

    – find(), find_all() 함수를 이용하여 스크레이핑

• def	scraping_use_select(html)
    
    – select(), select_one() 함수를 이용하여 스크레이핑

```python
# 모듈 설정
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def main(tool):
    # urlopen으로 링크를 html 파일 화
    html = urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168')
    # BeautifulSoup을 활용하여 잘못된 HTML을 수정하여 쉽게 탐색할 수 있는 XML 형식의 파이썬 객체로 변환
    soup = BeautifulSoup(html.read(), 'html.parser')
    # 다른 img등이 선택되는걸 방지하기위해 id가 seven-day-forcast-continer인 <div>태그를 따로 저장
    soup_seven_day = soup.select_one('div#seven-day-forecast-container')
    # find를 활용하는 경우
    # soup_seven_day = soup.find('div', {'id' : 'seven-day-forecast-container'})

    print('National Weather Service Scraping')
    print('----------------------------------')
    print(f'[{tool} 함수 이용]')

    if tool == 'find' :
        scraping_use_find(soup_seven_day)
    elif tool == 'select' :
        scraping_use_select(soup_seven_day)
```

```python
# find 함수 활용
def	scraping_use_find(soup_seven_day):
    list_period = soup_seven_day.find_all('p', attrs={'class' : 'period-name'}) # class가 period-name인 <p>태그를 리스트로 모두 저장
    list_short_desc = soup_seven_day.find_all('p', attrs={'class' : 'short-desc'}) # class가 .short-desc인 <p>태그를 리스트로 모두 저장
    list_temp = soup_seven_day.find_all('p', attrs={'class' : 'temp'}) # class가 temp인 <p>태그를 리스트로 모두 저장
    list_image_desc = soup_seven_day.find_all('img') # img 태그와 관련된걸 리스트로 모두 저장
    
    showscraping(list_period, list_short_desc, list_temp, list_image_desc)
```

```python
# select 함수 활용
def scraping_use_select(soup_seven_day):
    list_period = soup_seven_day.select('p.period-name') # class가 period-name인 <p>태그를 리스트로 모두 저장
    list_short_desc = soup_seven_day.select('p.short-desc') # class가 .short-desc인 <p>태그를 리스트로 모두 저장
    list_temp = soup_seven_day.select('p.temp') # class가 temp인 <p>태그를 리스트로 모두 저장
    list_image_desc = soup_seven_day.select('img') # img  태그와 관련된걸 리스트로 모두 저장

    showscraping(list_period, list_short_desc, list_temp, list_image_desc)
```

```python
# 출력 함수
def showscraping(list_period, list_short_desc, list_temp, list_image_desc):
    print('--------------------------------------------------------------------------')

    for i in range(0,len(list_period)):
        period = list_period[i].text # list_period의 i번째 요소를 text로 저장
        short_desc = list_short_desc[i].text # list_short_desc의 i번째 요소를 text로 저장
        temperature = list_temp[i].text # list_temp의 i번째 요소를 text로 저장
        image_desc = list_image_desc[i]['title'] # list_image_desc의 i번째 요소에서 'alt'에 해당하는 부분을 저장

        # 결과 출력
        print(f'[Period]: {period}\n[Short desc]: {short_desc}\n[Temperature]: {temperature}\n[Image desc]: {image_desc}')
        print('--------------------------------------------------------------------------')
```

```python
main('find')
```
