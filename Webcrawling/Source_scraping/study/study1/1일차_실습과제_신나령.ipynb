# 과제 내용
   - 일주일간의 일기예보 내용을 화면에 출력
   - 총 8일간의 일기예보 내용을 스크레이핑
   
   - 출력 항목 (4개 항목)  
        – period-name  
        – short-desc  
        – temp  
        – img title='...'  

```python
# 필요한 모듈
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

## – select(), select_one() 함수를 이용하여 스크레이핑

```python
html=urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytj6nXZBxPY') # 특정 웹사이트 주소
def scraping_use_select(html):
    bs=BeautifulSoup(html.read(),'html.parser')
    
    period_names=bs.select('.tombstone-container > .period-name')
    short_descs=bs.select('.tombstone-container > .short-desc')
    temps=bs.select('.tombstone-container > .temp')
    imgs=bs.select('.forecast-icon')
    
    print('National Weather Service Scraping\n----------------------------------\n[select 함수 이용]')
    print('--------------------------------------------------------------------------------')
    for i in range(9):
        print('[Period]: {}'.format(period_names[i].text))
        print('[Short desc]: {}'.format(short_descs[i].text))
        print('[Temperature]: {}'.format(temps[i].text))
        print('[Image desc]: {}'.format(imgs[i]['title']))
        print('--------------------------------------------------------------------------------')
scraping_use_select(html)
```

## – find(), find_all() 함수를 이용하여 스크레이핑

```python
html=urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytj7unZBxPZ') # 특정 웹사이트 주소
def scraping_use_find(html):
    bs=BeautifulSoup(html.read(),'html.parser')

    period_names=bs.find_all('p',{'class':'period-name'})
    short_descs=bs.find_all('p',{'class':'short-desc'})
    temps=bs.find_all('p',{'class':'temp'})
    imgs=bs.find_all('img',{'class':'forecast-icon'})
    
    print('National Weather Service Scraping\n----------------------------------\n[find 함수 이용]')
    print('--------------------------------------------------------------------------------')
    for i in range(9):
        print('[Period]: {}'.format(period_names[i].get_text()))
        print('[Short desc]: {}'.format(short_descs[i].get_text()))
        print('[Temperature]: {}'.format(temps[i].get_text()))
        print('[Image desc]: {}'.format(imgs[i].attrs['title']))
        print('--------------------------------------------------------------------------------')
scraping_use_find(html)
```
