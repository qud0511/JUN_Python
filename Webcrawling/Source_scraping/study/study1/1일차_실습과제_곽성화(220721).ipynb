# 과제
### 일주일간의 일기예보 내용을 화면에 출력

### 2개의 함수를 작성
출력 내용은 동일하지만, 아래 2개의 함수를 각각 구현하시오.
   * def scraping_use_find(html)   
        – find(), find_all() 함수를 이용하여 스크레이핑
   * def scraping_use_select(html)   
        – select(), select_one() 함수를 이용하여 스크레이핑

### 참고
1. div class="tombstone-container" 찾기
2. 4개 항목 출력   
 period-name / short-desc / temp / img title=“...”

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

# find() 함수 이용
def scraping_use_find(html):
    page = BeautifulSoup(urlopen(html).read(), 'html.parser')
    seven_day=page.find('div',{'id':'seven-day-forecast'})     
    forecast_item = seven_day.find_all('div',{'class':'tombstone-container'})
    
    print('[find 함수 이용]')
    for item in forecast_item:
        print('-'*70)
        print(f"[period]: {item.find('p',{'class':'period-name'}).text}")
        print(f"[Short desc]: {item.find('p',{'class':'short-desc'}).text}")
        print(f"[Temperature]: {item.find('p',{'class':'temp'}).text}")
        print(f"[Image desc]: {item.find('img')['title']}")

        
# select() 함수 이용
def scraping_use_select(html):
    page = BeautifulSoup(urlopen(html).read(), 'html.parser')
    seven_day=page.select_one('div#seven-day-forecast')
    forecast_item = seven_day.select('div.tombstone-container')
    
    print('[select 함수 이용]')
    for item in forecast_item:
        print('-'*70)
        print(f"[period]: {item.select_one('p.period-name').text}")
        print(f"[Short desc]: {item.select_one('p.short-desc').text}")
        print(f"[Temperature]: {item.select_one('p.temp').text}")
        print(f"[Image desc]: {item.select_one('img')['title']}")
```

```python
html='https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YtjtEXZBy3A'
scraping_use_find(html)
print('\n'+'='*70)
scraping_use_select(html)
```
