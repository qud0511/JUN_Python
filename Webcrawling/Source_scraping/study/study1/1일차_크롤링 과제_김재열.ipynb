## select 함수 선언 및 호출

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

def scraping_use_select(url):
    html=urlopen(url)
    bs= BeautifulSoup(html.read(), 'html.parser')
    periods=bs.select('p.period-name')
    desc=bs.select('p.short-desc')
    temp = bs.select('p.temp')
    for i in range(len(bs.select('div.tombstone-container'))):
        print(f'[Period] {periods[i].text}')
        print(f'[Short desc] {desc[i].text}')
        print(f'[Temperature] {temp[i].text}')
        print(f"[Image desc] {img_title[i].select_one('img')['title']}")
        print()
        print('-'*50)
```

```python
scraping_use_find('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YsuIHi9yxTZ')
```
