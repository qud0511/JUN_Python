```python
# MODULE
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytk4nHZBxEY')
soup=BeautifulSoup(html.read(), 'html.parser') 
```

```python
sevendays=soup.find('div', {'id':'seven-day-forecast'})
print(sevendays)
```

```python
# 출력 항목 (4개 항목)
    # – period-name
    # – short-desc
    # – temp
    # – <img title=“...”>
```

### find & find_all :

```python
weekdays=sevendays.find('div', {'class':"tombstone-container"})
periods=weekdays.find('p', {'class':'period-name'}).text
short_d=weekdays.find('p',{'class':'short-desc'}).text
temps_high=weekdays.find('p',{'class':'temp temp-high'}).text
# temps_low=weekdays.find('p',{'class':'temp temp-low'}).text
image=weekdays.find('img', {'class':'forecast-icon'})
image_d=image.get('alt')

print(weekdays, '\n')
print(periods, '\n')
print(short_d, '\n')
print(temps_high, '\n')
# print(temps_low, '\n')
print(image_d)
```

```python
def scraping_use_find(sevendays):
    head='''National Weather Service Scraping
----------------------------------
[find 함수 이용]
--------------------------------------------------------------------------------'''
    print(head)
    weekdays_all=sevendays.find_all('div', {'class':"tombstone-container"})
    
    for weekday in weekdays_all:
        periods=weekday.find('p', {'class':'period-name'}).text
        short_d=weekday.find('p',{'class':'short-desc'}).text
        temps=weekday.find('p',{'class':'temp temp-high'})
        if temps==None:
            temps=weekday.find('p',{'class':'temp temp-low'})
        image=weekday.find('img', {'class':'forecast-icon'})
        image_d=image.get('alt')

        print('[Period]: ', periods)
        print('[Short desc]:', short_d)
        print('[Temperature]:', temps.text)
        print('[Image desc]: ', image_d)
        print('--------------------------------------------------------------------------------')

scraping_use_find(sevendays)
```

### select & select_one :

```python
def scraping_use_select(sevendays):
    head='''National Weather Service Scraping
----------------------------------
[select 함수 이용]
--------------------------------------------------------------------------------'''
    print(head)
    weekdays_all=sevendays.select('.tombstone-container')
    for weekday in weekdays_all:
        periods=weekday.select_one('.period-name').text
        short_d=weekday.select_one('.short-desc').text
        temps=weekday.select_one('.temp').text           # why not '.temp temp' high b/ '.temp'?
        image=weekday.select_one('.forecast-icon')['title']     # 

        print('[Period]: ', periods)
        print('[Short desc]:', short_d)
        print('[Temperature]:', temps)
        print('[Image desc]: ', image)
        print('--------------------------------------------------------------------------------')

scraping_use_select(sevendays)
```
