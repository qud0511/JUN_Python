### 2. 저장된 csv파일을 읽어서 사용자가 입력한 지역에 있는 매장 정보 출력

```python
# MODULE
import pandas as pd
import re

# FILE
FILE='./hollys_branches.csv'

# READ
branchesDF=pd.read_csv(FILE, encoding='utf-8', index_col=0)
```

```python
branchesDF.columns=['지역', '매장명', '주소', '전화번호']
branchesDF.sort_values('지역', inplace=True)
```

```python
branchesDF.head()
```

```python
# index 정리
branchesDF.index=range(len(branchesDF.index))
branchesDF
```

```python
# input
test=False
if test == True:
    while True:
        keyword=input('검색할 매장의 도시를 입력하세요: ').split()
        if len(keyword)==2:
            break
        else:
            print('--> 서울 종로 <-- 와 같이 입력해주세요.')
```

```python
while True:
        keyword=input('검색할 매장의 도시를 입력하세요: ').split()
        if len(keyword)==2:
            break
        else:
            print('--> 서울 종로 <-- 와 같이 입력해주세요.')
result=[]
for i in branchesDF.index:
    location=branchesDF.iloc[i]['지역']
    if keyword[0] in location and keyword[1] in location:
        result.append(list(branchesDF.iloc[i]))
if len(result)==0:
    print(f'{keyword[0]} {keyword[1]} 지역 매장이 존재하지 않습니다')
else:
    print(f'''--------------------
검색된 매장 수: {len(result)}
--------------------''')
    for i in range(len(result)):
        print(f'[{i:>2d}]: {result[i][2:]}')

```
