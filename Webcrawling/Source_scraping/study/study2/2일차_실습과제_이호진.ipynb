```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests

html = urlopen('https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=10&sido=&gugun=&store=')
bs = BeautifulSoup(html.read(), 'html.parser')
print(bs)
```

```python
cafe_place=[]
cafe_name=[]
cafe_address=[]
cafe_num=[]
for i in range(1,54):
    print(f'{i}페이지 분석중....')
    html = urlopen(f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={i}&sido=&gugun=&store=')
    bs = BeautifulSoup(html.read(), 'html.parser')
    bs2 = bs.find_all("td")
    # 필요한 데이터만 추출(인덱싱 사용)
    for j in range(0, len(bs2), 6):
        cafe_place.append(bs2[j].text)
        cafe_name.append(bs2[j+1].text)
        cafe_address.append(bs2[j+3].text)
        cafe_num.append(bs2[j+5].text)
        
```

```python
import pandas as pd
coffeeDF=pd.DataFrame({'매장이름':cafe_name, '위치':cafe_place, '주소':cafe_address, '전화번호':cafe_num})
coffeeDF.to_csv('hollys_branches.csv', encoding='cp949')
```
