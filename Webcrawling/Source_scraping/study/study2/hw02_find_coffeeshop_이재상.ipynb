```python
# 모듈 로딩
import pandas as pd
import csv
```

```python
# DF 생성
f = open('hollys_branches.csv', encoding = 'utf-8')
data = csv.reader(f)

result = []
count = 0

# 검색 및 결과 출력 함수
search = input('검색할 매장의 도시를 입력하세요 : ')
print('검색할 매장의 도시를 입력하세요 : {0}'.format(search))
print('-'*80)

for row in data:
    if search in row[1]:
        for data in row[3:]:
            result.append(data)

print('검색된 매장 수 : {0}개'.format(round(len(result)/2)))
print('-'*80)

for n in range(0,len(result),2):
    num = result[n:n+2]
    count+=1
    print('[{0}] : {1}'.format(count,num))
```
