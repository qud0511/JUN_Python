```python
import pandas as pd
import re

def find_store(city):
    hollysDF=pd.read_csv('hollys_branches.csv',encoding='utf-8')
    store=[]
    for i in range(len(hollysDF)):
        if city in hollysDF['위치(시,구)'][i]:
            store.append(hollysDF.iloc[i])
    storeDF=pd.DataFrame(store)
    print('검색된 매장 수: {}'.format(len(storeDF)))
    print('--------------------')
    for i in range(len(storeDF)):
        print("[{}]: ['{}', '{}']".format(i+1,storeDF['주소'].to_list()[i],storeDF['전화번호'].to_list()[i]))
city=input('검색할 매장의 도시를 입력하세요: ')     
find_store(city)
```
