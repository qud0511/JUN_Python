```python
import pandas as pd
df=pd.read_csv('hollys_branches.csv')
df.head(20)
```

```python
city=input('검색할 매장의 도시를 입력하세요: ').split()
list=[]
for i in df.index:
    add=df['주소'].str.split()
    if (city[0] in add[i][0]) and (city[1] in add[i][1]):
        list.append(df.iloc[i])

result=pd.concat(list, axis=1, ignore_index=True)
result.drop('Unnamed: 0', inplace=True)
result=result.T
print('-'*20)
print(f'검색된 매장 수: {result.shape[0]}')
print('-'*20)

for i in result.index:
    print('[ {0}]: [{1}, {2}]'.format(i+1, result.iloc[i]['주소'], result.iloc[i]['전화번호']))
```
