```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

# ---------------------------------------------
# input : 페이지 number
# output : 해당 페이지내 매장 정보 데이터 프레임
# ---------------------------------------------
def add_center_page(page_num) :
    HOLLYS_STORE_URL = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_num}&sido=&gugun=&store='
    html = urlopen(HOLLYS_STORE_URL)
    soup = BeautifulSoup(html, 'html.parser')
    
#     표를 담은 tbody와 그 안의 tr 들을 읽어오기
    tbody = soup.find('tbody')
    trs = tbody.select('tr')
    
#     매장의 정보를 저장할 list
    center_city_list, center_name_list, center_loc_list, center_call_list  = [], [], [],[]
    
#     표 내의 10개의 행에 대해서 반복, 리스트에 정보 추가
    for tr in trs:
        center_city_list.append(tr.find('td', class_ = 'noline center_t').text)
        center_name_list.append(tr.select_one('td > a').text)
        center_loc_list.append(tr.select('td > a')[1].text)
        center_call_list.append(tr.select('td')[-1].text)
    
#     해당 페이지 내의 매장에대한 정보를 담은 데이터 프레임 생성
    center_df_page = pd.DataFrame({
        '매장이름':center_name_list, 
        '위치(시,구)': center_city_list,
        '주소': center_loc_list,
        '전화번호': center_call_list
        })
    
    return center_df_page
    
```

```python
# 빈 데이터 프레임 생성
total_center = pd.DataFrame(columns = add_center_page(1).columns)


for i in range(1, 54):
    print(f'{i} page 읽어 오는중...')
    
#     1 ~ 53 페이지의 데이터 프레임을 추가
    total_center = pd.concat([total_center, add_center_page(i)]).reset_index(drop = True)
    
total_center
```

```python
total_center.to_csv('hollys_branches.csv', encoding='utf-8', index=False)
```
