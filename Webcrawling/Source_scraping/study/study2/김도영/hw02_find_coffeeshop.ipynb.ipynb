```python
# 모듈 로딩
import pandas as pd

# 파일 로딩
df=pd.read_csv('hollys_branches.csv', encoding='utf-8')
df
```

```python
# 'Unnamed: 0' 컬럼 삭제 
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
# 데이터 프레임 확인
df
```

```python
# 값 입력받기
n=input('검색할 매장의 도시를 입력하세요: ').split()

# 값에 대한 행 추출
value=[]
for i in df.index:
    sr=df['주소'].str.split()
    if (n[0] in sr[i][0]) and (n[1] in sr[i][1]):
        value.append(df.iloc[i])

# 추출한 행 데이터 프레임 형성
df2=pd.concat(value,axis=1, ignore_index=True)
df2=df2.T
df2

# 결과 출력
print('--------------------')
print(f'검색된 매장 수: {df2.shape[0]}')
print('--------------------')

for i in df2.index:
    print("[{:>3}]: ['{}', '{}']".format(i+1,df2.iloc[i]['주소'],df2.iloc[i]['전화번호']))
```
