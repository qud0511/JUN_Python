```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup

html=requests.get('https://finance.naver.com/sise/sise_market_sum.naver')
soup=BeautifulSoup(html.text,'html.parser')
stocks=soup.find('div',class_="box_type_l").find_all('td')
stock_list=[]
stock_url=[]
stock_dict={}

for stock in stocks:
    if len(stock('a'))!=0:
        if stock('a')[0].text!='':
            stock_url.append(stock('a')[0]['href'])
            stock_list.append(stock('a')[0].text)
        if len(stock_list)==10: break
print('-------------------------------------\n[ 네이버 코스피 상위 10대 기업 목록 ]\n-------------------------------------')
for i in range(len(stock_list)):
    print('[{0:2}] {1}'.format(i+1,stock_list[i]))
    stock_dict[i+1]=stock_url[i]   
while True:
    num=input('주가를 검색할 기업의 번호를 입력하세요(-1: 종료)')
    if int(num)==-1:
        print('프로그램 종료')
        break
    elif 1<=int(num)<=10:
        url='https://finance.naver.com/'+stock_dict.get(int(num))
        print(url)

        html=requests.get(url)
        soup=BeautifulSoup(html.text,'html.parser')
        item_list=soup.find('dl').find_all('dd')
        total_list=[]

        for item in item_list:
            total_list.append(item.text)

        for i in range(len(total_list)):
            if i in (1,2,3,4,5,6,8):
                print('{}: {}'.format(total_list[i].split(' ')[0],total_list[i].split(' ')[1]))
    else: continue
    
```
