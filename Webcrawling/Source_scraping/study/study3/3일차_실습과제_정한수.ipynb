```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
head_url = 'https://finance.naver.com/'
url = head_url+'/sise/sise_market_sum.naver'
```

```python
def print_stock_list(stock_list):
    print('\n-------------------------------------')
    print('[ 네이버 코스피 상위 10대 기업 목록 ]')
    print('-------------------------------------')
    for stock in stock_list:
        print('[{0:2} {1}]'.format(stock[0], stock[1]))

def get_stock_detail(detail_url):
    html = requests.get(detail_url)
    soup = BeautifulSoup(html.text, 'html.parser')
    
    code = soup.select_one('div.description > span').text
    rate_info = soup.find('div', class_ = 'rate_info')
    curr = rate_info.find('p', class_ = 'no_today').select_one('span.blind').text
    rate_info = rate_info.select('tr')
    tr1, tr2 = rate_info[0], rate_info[1]
    yesterday = tr1.find('td', class_ = 'first').select_one('span.blind').text
    siga = tr2.find('td', class_="first").select_one('span.blind').text
    high = tr1.select('td')[1].select_one('span.blind').text
    low = tr2.select('td')[1].select_one('span.blind').text
    
    print('종목코드:', code)
    print('현재가:', curr)
    print('전일가:', yesterday)
    print('시가:', siga)
    print('고가:', high)
    print('저가:', low, )
    

while True:
    # 주가는 실시간으로 변하므로 출력 직전 매번 최신화
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    stock_list = []

    trs = soup.find_all('tr', {"onmouseover":"mouseOver(this)"})[:10]
    for tr in trs:
        rank = tr.find('td', class_ = 'no').text
        title = tr.find('a', class_ = 'tltle')
        href = title.attrs['href']
        name = title.text

        stock_list.append([rank, name, href])


    print_stock_list(stock_list)
    stock_num = int(input('주가를 검색할 기업의 번호를 입력하세요(-1: 종료): '))

    if stock_num == -1:
        print('프로그램 종료')
        break
    elif stock_num < 1 or stock_num >11:
        continue
    
    detail_url = head_url+stock_list[stock_num-1][-1]
    print('\n'+detail_url)
    print('종목명:', stock_list[stock_num-1][1])
    get_stock_detail(detail_url)
    
```
