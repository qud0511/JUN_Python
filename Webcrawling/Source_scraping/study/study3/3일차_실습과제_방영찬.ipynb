```python
import requests
from bs4 import BeautifulSoup

html = requests.get('https://finance.naver.com/sise/sise_market_sum.naver')
soup = BeautifulSoup(html.text, 'html.parser')
url_main = 'https://finance.naver.com'

titles=soup.select('tbody > tr > td > a.tltle')

n = 0

url_list=[]
title_list=[]

for i in range(len(titles)):
    url_list.append(titles[i]['href'])
for title in titles:
    title_list.append(title.text)

print('-'*30)
print('[ 네이버 코스피 상위 10대 기업 목록 ]')
print('-'*30)

for i in range(10):
    print('[{0:>2}] {1}'.format(i+1, title_list[i]))
for i in range(10):
    globals()["html_detail_{}".format(i)] = requests.get(url_main + url_list[i])
    globals()["soup_detail_{}".format(i)] = BeautifulSoup(globals()["html_detail_{}".format(i)].text, 'html.parser')
    globals()["detail_list_{}".format(i)] = globals()["soup_detail_{}".format(i)].select('div > dl > dd')


try:
    while n != -1:
        n = int(input('주가를 검색할 기업의 번호를 입력하세요(-1: 종료): '))                                        
        print(url_main + url_list[n-1])
        print(globals()["detail_list_{}".format(n-1)][1].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][1].text.split()[1])
        print(globals()["detail_list_{}".format(n-1)][2].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][2].text.split()[1])
        print(globals()["detail_list_{}".format(n-1)][3].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][3].text.split()[1])
        print(globals()["detail_list_{}".format(n-1)][4].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][4].text.split()[1])
        print(globals()["detail_list_{}".format(n-1)][5].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][5].text.split()[1])
        print(globals()["detail_list_{}".format(n-1)][6].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][6].text.split()[1])
        print(globals()["detail_list_{}".format(n-1)][8].text.split()[0] + ': '
            + globals()["detail_list_{}".format(n-1)][8].text.split()[1])
except:
    print('검색을 종료합니다.')
```
