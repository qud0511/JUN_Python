```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import requests
from itertools import count
import time

html = requests.get('https://finance.naver.com/sise/sise_market_sum.naver')
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
finance_num = ['005930', '373220', '000660', '207940', '005935', '005380', '035420', '051910', '006400', '000270']

def get_topten_data():
    print('-----------------------------------')
    print('[ 네이버 코스피 상위 10대 기업 목록 ]')
    print('-----------------------------------')
    finance_list = []
    for title in soup.select('div.box_type_l > table > tbody > tr > td > a'):
        if title.text !='':
            finance_list.append(title.text)
    for i in range(0, 10):
        print('[ {}] {}'.format(i+1, finance_list[i]))
    print(' ')


def get_finance_data(find):
    wd = webdriver.Chrome('C:/Users/User/chromedriver_win32/chromedriver.exe')
    financeURL = 'https://finance.naver.com/item/main.naver?code=%s' %finance_num[int(find)-1]
    wd.get(financeURL)
    time.sleep(5)

    information = []   

    html = wd.page_source
    soupData = BeautifulSoup(html, 'html.parser')
    finance_info= list(soupData.select('.new_totalinfo > dl > dd'))

    finance_info_list = []

    for info in finance_info:
        finance_info_list.append(info.text)

    nameData = finance_info_list[1][4:]
    codeData=finance_info_list[2][5:11]
    today=finance_info_list[3][4:10]
    yesterday=finance_info_list[4][4:10]
    market_price=finance_info_list[5][3:10]
    max_price=finance_info_list[6][3:10]
    min_price=finance_info_list[7][4:10]

    information.append([nameData, codeData, today, yesterday,market_price, max_price, min_price])


    print(financeURL)
    print('종목명:', information[0][0])
    print('종목코드:', information[0][1])
    print('현재가:', information[0][2])
    print('전일가:', information[0][3])
    print('시가:', information[0][4])
    print('고가:', information[0][5])
    print('저가:', information[0][6])

def main():
    while True:
        get_topten_data()
        find = input('주가를 검색할 기업의 번호를 입력하세요(-1: 종료):')
        if (int(find) == -1): 
            print('프로그램 종료')
            break
        else: get_finance_data(find)

    
main()
```
