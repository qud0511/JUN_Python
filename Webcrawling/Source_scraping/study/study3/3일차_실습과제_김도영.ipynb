```python
# 모듈 로딩
import requests
from bs4 import BeautifulSoup

# 전체 파일 필요한 값 리스트에 담기
html=requests.get('https://finance.naver.com/sise/sise_market_sum.naver')
soup=BeautifulSoup(html.text, 'html.parser')

tbody=soup.find('tbody')
a=tbody.find_all('a')

value=[]  
for i in a[0:20:2]:
    value.append(i)

URL=[]  # 주소 담는 리스트
company=[]  # 기업명 담는 리스트

for i in value:
    URL.append(i['href'])
    company.append(i.text)
    
# print(URL)
# print(company)
```

```python
# 실행 결과
while True:
    print('-------------------------------------')
    print('[ 네이버 코스피 상위 10대 기업 목록 ]')
    print('-------------------------------------')
    for i in range(len(company)):
        print('[{:>2}] {}'.format(i+1,company[i]))    
    
    n=int(input('주가를 검색할 기업의 번호를 입력하세요(-1: 종료):'))
    
    if n>=1 and n<=10:
        # 해당 페이지 필요한 값 리스트에 담아 출력하기
        html1=requests.get('https://finance.naver.com'+URL[n-1])
        soup1=BeautifulSoup(html1.text, 'html.parser')

        middle=soup1.find('div',{'id':'middle'})
        value_2=[]

        for i in middle.find_all('dd')[1:9]:
            value_2.append(i.text)

        item=[] # 종목명, 종목코드, 현재가, 전일가, 시가, 고가, 저가 
        result=[] # (종목명, 종목코드, 현재가, 전일가, 시가, 고가, 저가)의 값 담을 리스트
        for i in value_2:
            item.append(i.split()[0])
            result.append(i.split()[1])

        del item[6]
        del result[6]

        print('https://finance.naver.com'+URL[n-1])
        for i in range(7):
            print('{} : {}'.format(item[i],result[i]))

    if n==-1: 
        # 프로그램 종료하기
        print('프로그램 종료')
        break
        
    elif n<1 or n>10:
        # 잘못된 번호 다시 입력 받기
        print('잘못된 번호 입니다. 다시 입력해 주세요.')
```
