```python
url = 'https://finance.naver.com/sise/sise_market_sum.naver'
html = requests.get(url)
bs = BeautifulSoup(html.text, 'html.parser')
bs
```

```python
bs1 = bs.find_all('tr')
bs1
```

```python
linkList = []
nameList = []
for i in range(7,12):
    linkList.append(bs1[i].find('a')['href'])
    nameList.append(bs1[i].find('a',class_='tltle').text)
for j in range(15,20):
    linkList.append(bs1[j].find('a')['href'])
    nameList.append(bs1[j].find('a',class_='tltle').text)
linkList
```

```python
nameList
```

```python
url_1 = 'https://finance.naver.com'
info_list=[]
url_list=[]
for link in linkList:
    html1 = requests.get(url_1+link)
    bs1 = BeautifulSoup(html1.text, 'html.parser')
    bs2 = bs1.find_all('dd')
    for i in range(1,9):
        if i == 7:
            continue
        else:
            info_list.append(bs2[i].text.split()[1])
    
    url_list.append(url_1+link)
info_list
```

```python
url_list
```

```python
print(f"{'*'*30}")
print('[ 네이버 고스피 상위 10대 기업 목록 ]')
print(f"{'*'*30}")

for i in range(1,len(nameList)+1):
    print(f'[{i:2}] {nameList[i-1]}')

while True:
    num = int(input('추가를 검색할 기업의 번호를 입력하세요(-1: 종료): '))
    if num >=1 and num <=10:
        print(url_list[num-1])
        print(f'종목명: {info_list[num-1]}')
        print(f'종목코드: {info_list[num]}')
        print(f'현재가: {info_list[num+1]}')
        print(f'전일가: {info_list[num+2]}')
        print(f'시가: {info_list[num+3]}')
        print(f'고가: {info_list[num+4]}')
        print(f'저가: {info_list[num+5]}')
    elif num == -1:
        print('프로그램 종료')
        break
```
